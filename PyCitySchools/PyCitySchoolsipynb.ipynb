```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# Files to Load 
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames as follows
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name"])
```

```python
# Calculate school count, student count, and the total budget as follow:
school_count = len(school_data_complete["school_name"].unique())
student_count = school_data_complete["Student ID"].count()
total_budget = school_data["budget"].sum()

# Calculate the average maths and reading scores as follows:
average_maths_score = school_data_complete["maths_score"].mean()
average_reading_score = school_data_complete["reading_score"].mean()

# Calculate the number of students passing both math and reading (passing grade is above or equal to 50):
passing_math_count = school_data_complete[(school_data_complete["maths_score"] >= 50)].count()["student_name"]
passing_reading_count = school_data_complete[(school_data_complete["reading_score"] >= 50)].count()["student_name"]

# Calculate the percentage of studnets passing maths and reading as follows:
passing_math_percentage = (passing_math_count / student_count) * 100
passing_reading_percentage = (passing_reading_count / student_count) * 100

# Total number of students passing both maths and reading, and find the percentage of that number as follows:
passing_math_reading_count = school_data_complete[(school_data_complete["maths_score"] >= 50) & (school_data_complete["reading_score"] >= 50)].count()["student_name"]
passing_math_reading_percentage = (passing_math_reading_count / student_count) * 100

# Create a "government summary" data frame with the following columns:
local_government_summary_df = pd.DataFrame({
    "Total Schools" : school_count, 
    "Total Students" : student_count, 
    "Total Budget" : total_budget,
    "Average Maths Score" : average_maths_score, 
    "Average Reading Score" : average_reading_score, 
    "% Passing Maths" : passing_math_percentage, 
    "% Passing Reading" : passing_reading_percentage, 
    "% Overall Passing" : passing_math_reading_percentage
}, index=[0])


# Change the format of both "Total Studnets" and "Total Budget" columns as follows:
local_government_summary_df["Total Students"] = local_government_summary_df["Total Students"].map("{:,.0f}".format)
local_government_summary_df["Total Budget"] = local_government_summary_df["Total Budget"].map("${:,.2f}".format)

# Print the data frame
local_government_summary_df
```

```python
# Set index to "school_name":
school_type = school_data.set_index(["school_name"])["type"]

# Find number of studnets per school:
school_student_count = school_data_complete["school_name"].value_counts()

# Find the budget for each school:
school_budget = school_data_complete.groupby(["school_name"]).mean()["budget"]

# Find budget per student:
per_student_budget = school_budget / school_student_count

# Calculate average scores per school for maths and reading
average_maths_score = school_data_complete.groupby(["school_name"]).mean()["maths_score"]
average_reading_score = school_data_complete.groupby(["school_name"]).mean()["reading_score"]

# Calculate number of students passing maths and reading:
maths_passing = school_data_complete[(school_data_complete["maths_score"]) >= 50]
reading_passing = school_data_complete[(school_data_complete["reading_score"]) >= 50]

# Calculate total number of students passing per school as follow:
school_maths_passing = maths_passing.groupby(["school_name"]).count()["Student ID"]
school_reading_passing = reading_passing.groupby(["school_name"]).count()["Student ID"]

# Calculate percentage values as follows:
maths_percentage = (school_maths_passing / school_student_count) * 100
reading_percentage = (school_reading_passing / school_student_count) * 100

# Calculate total number of students per school who are passing both maths and reading:
maths_reading_passing = school_data_complete[(school_data_complete["maths_score"] >= 50) & (school_data_complete["reading_score"] >= 50 )]

maths_reading_passing_count = maths_reading_passing.groupby(["school_name"]).count()["Student ID"]

maths_reading_passing_perentage = (maths_reading_passing_count / school_student_count) * 100

# Create a new "schools summary" data frame as follows:
local_schools_summary_df = pd.DataFrame({
    "School Type" : school_type, 
    "Total Students" : school_student_count, 
    "Total School Budget" : school_budget,
    "Per Student Budget" : per_student_budget, 
    "Average Maths Score" : average_maths_score, 
    "Average Reading Score" : average_reading_score, 
    "% Passing Maths" : maths_percentage,
    "% Passing Reading" : reading_percentage, 
    "% Overall Passing" : maths_reading_passing_perentage
})

# Change the format of both "Total School Budget" and "Total Student Budget" columns as follows:
local_schools_summary_df["Total School Budget"] = local_schools_summary_df["Total School Budget"].map("${:,.2f}".format)
local_schools_summary_df["Per Student Budget"] = local_schools_summary_df["Per Student Budget"].map("${:,.2f}".format)

# Print data frame:
local_schools_summary_df
```

```python
# Find highest performing schools in "% Overall Passing":
highest_schools = local_schools_summary_df.sort_values(["% Overall Passing"], ascending=False)
highest_schools.head(5)
```

```python
# Use booleans to create the following grade lists:
ninth_grade = school_data_complete[(school_data_complete["year"] == 9)]
tenth_grade = school_data_complete[(school_data_complete["year"] == 10)]
eleventh_grade = school_data_complete[(school_data_complete["year"] == 11)]
twelvth_grade = school_data_complete[(school_data_complete["year"] == 12)]

# Find average maths scores for each grade per different schools as follows:
ninth_grade_math = ninth_grade.groupby(["school_name"]).mean()["maths_score"]
tenth_grade_math = tenth_grade.groupby(["school_name"]).mean()["maths_score"]
eleventh_grade_math = eleventh_grade.groupby(["school_name"]).mean()["maths_score"]
twelvth_grade_math = twelvth_grade.groupby(["school_name"]).mean()["maths_score"]

# Create a new "maths scores" data frame as follow:
maths_scores_df = pd.DataFrame({
    "Year 9" : ninth_grade_math, 
    "Year 10" : tenth_grade_math, 
    "Year 11" : eleventh_grade_math, 
    "Year 12" : twelvth_grade_math
})

# Remove the index from the data frame:
maths_scores_df.index.name = None

# Print the data frame:
maths_scores_df
```

```python
# Find average reading scores for each grade per different schools as follows:
ninth_grade_reading = ninth_grade.groupby(["school_name"]).mean()["reading_score"]
tenth_grade_reading = tenth_grade.groupby(["school_name"]).mean()["reading_score"]
eleventh_grade_reading = eleventh_grade.groupby(["school_name"]).mean()["reading_score"]
twelvth_grade_reading = twelvth_grade.groupby(["school_name"]).mean()["reading_score"]

# Create a new "reading scores" data frame as follow:
reading_scores_df = pd.DataFrame({
    "Year 9" : ninth_grade_reading, 
    "Year 10" : tenth_grade_reading, 
    "Year 11" : eleventh_grade_reading, 
    "Year 12" : twelvth_grade_reading
})

# Remove the index from data frame:
reading_scores_df.index.name = None

# Print the data frame:
reading_scores_df
```

```python
# Establish the spending bins and group names as follows:
spending_bins = [0, 585.9, 615.9, 645.9, 675.9]

group_names = ["<$585", "$585-630", "$630-645", "$645-680"]

 
# Categorize spending based on the bins
local_schools_summary_df["Spending Ranges (Per Student)"] = pd.cut(per_student_budget, spending_bins, labels=group_names)

# Find average maths and reading scores by grouping values on "Spending Ranges (Per Student)" for average scores:
spending_maths_score = local_schools_summary_df.groupby(["Spending Ranges (Per Student)"]).mean()["Average Maths Score"]
spending_reading_score = local_schools_summary_df.groupby(["Spending Ranges (Per Student)"]).mean()["Average Reading Score"]

# Find average maths and reading scores by grouping values on "Spending Ranges (Per Student)" for % passing values:
spending_maths_passing = local_schools_summary_df.groupby(["Spending Ranges (Per Student)"]).mean()["% Passing Maths"]
spending_reading_passing = local_schools_summary_df.groupby(["Spending Ranges (Per Student)"]).mean()["% Passing Reading"]

# Find average overall passing scores by grouping values on "Spending Ranges (Per Student)" for % Overall Passing values:
overall_passing = local_schools_summary_df.groupby(["Spending Ranges (Per Student)"]).mean()["% Overall Passing"]

# Create a new "school spending" data frame as follow:
school_spending_df = pd.DataFrame({
    "Average Maths Score" : spending_maths_score, 
    "Average Reading Score" : spending_reading_score, 
    "% Passing Maths" : spending_maths_passing, 
    "% Passing Reading" : spending_reading_passing, 
    "% Overall Passing" : overall_passing
    
})

# Change the format of the following columns as follows:
school_spending_df["Average Maths Score"] = school_spending_df["Average Maths Score"].map("{:,.2f}".format)
school_spending_df["Average Reading Score"] = school_spending_df["Average Reading Score"].map("{:,.2f}".format)
school_spending_df["% Passing Maths"] = school_spending_df["% Passing Maths"].map("{:,.2f}".format)
school_spending_df["% Passing Reading"] = school_spending_df["% Passing Reading"].map("{:,.2f}".format)
school_spending_df["% Overall Passing"] = school_spending_df["% Overall Passing"].map("{:,.2f}".format)


# Print the data frame:
school_spending_df

```
